```python
import pandas as pd
import pickle
```

```python
tablas = pd.read_html('https://web.archive.org/web/20240524150955/https://en.wikipedia.org/wiki/UEFA_Euro_2024')
```

```python
tablas[19]
```

```python
group_dict = {19: 'A', 26: 'B', 33: 'C', 40: 'D', 47: 'E', 54: 'F'}
dic_tablas={}

for i in range(19, 54, 7):
    group_letter = group_dict[i] 
    df = tablas[i]
    df.rename(columns={df.columns[1]: 'Team'}, inplace=True)
    df.pop('Qualification') 
    dic_tablas[f'Group {group_letter}'] = df
    print(f"Group {group_letter}:")
    print(df)
```

```python
dic_tablas['Group A']
```

```python
with open('dic_mundial', 'wb') as output:
    pickle.dump(dic_tablas, output)
```
